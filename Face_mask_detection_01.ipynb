```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
import albumentations as A
from albumentations.pytorch import ToTensorV2



def get_val_transform(image_size):
    val_aug_list=[
             A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
             A.Resize(image_size,image_size,always_apply=True),
             ToTensorV2(),
             ]
    return A.Compose(val_aug_list,p=1.)

image_transforms=get_val_transform(128)    
```

```python
import timm
import torch.nn as nn
class Efficient_model(nn.Module):
    
    def __init__(self, num_classes=2, model_name='tf_efficientnet_b0', pretrained=False):
        super(Efficient_model, self).__init__()
        self.model=timm.create_model(model_name, pretrained=pretrained, in_chans=3)
        self.model.classifier=nn.Linear(self.model.classifier.in_features, num_classes)

    def forward(self,x):
        x = self.model(x)
        return x

def load_checkpoint(path,model,device):
    
    checkpoint = torch.load(path,map_location=device)
    model.load_state_dict(checkpoint['model_state_dict'])
    print('...loading model ', end='')

    epoch = checkpoint['epoch']
    val_loss = checkpoint['val_loss']

    print(f'epoch={epoch}, val_loss={val_loss} ', end='')

    if  'f1_score' in checkpoint :
        f1_score=checkpoint["f1_score"]
        print(f'F1-score={f1_score}')    
```

```python
net=Efficient_model().to(device).eval()
load_checkpoint('/content/Eff_model_128_01.pth',net,device)
```

```python
import cv2
from IPython.display import clear_output
import time
from PIL import Image
from sort import Sort
import torch.nn.functional as F

%pylab inline 

id_label={0:'no mask',
          1:'mask'}

def get_detections(img_rgb,preds,Mask_model):
    labels=[]
    class_score=[]
    for conf,bbox in zip(preds['scores'],preds['boxes']):
        x1,y1,x2,y2=bbox
        conf=str(conf)[0:4]

        
        PIL_img=Image.fromarray(img_rgb)
        face_img=PIL_img.crop((x1,y1,x2,y2))
        face_img=np.array(face_img)
      


        img_tensor=image_transforms(image=face_img)['image'].to(device)
        
        output=Mask_model(img_tensor.unsqueeze(0))
        out_softmax=F.softmax(output,dim=1)
        id_=torch.argmax(out_softmax,dim=1).item()
        
 
        class_score.append(out_softmax[0][id_].item())
        labels.append(int(id_))

    detections=np.concatenate((preds['boxes'],\
                               np.array(preds['scores']).reshape(len(preds['scores']),-1),\
                               np.array(class_score).reshape(len(preds['scores']),-1),\
                               np.array(labels).reshape(len(preds['scores']),-1)), axis=1)
    return detections


def draw_bounding_boxes(img_bgr,tracked_objects):
    for x1, y1, x2, y2, obj_id,class_id in tracked_objects:

        label=id_label[int(class_id)]
        if int(class_id)==0:
          color=(0,0,255)
        else:
          color=(0,255,0)
        
        cv2.rectangle(img_bgr, (int(x1), int(y1)), (int(x2), int(y2)), color, 2)
        cv2.rectangle(img_bgr, (int(x1),int(y1)-35), (int(x1)+len(label)*19+60,int(y1)), color, -1)
        cv2.putText(img_bgr,label, (int(x1),int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,0), 3)
    return img_bgr

def mask_detection(video_path,FaceNet_model,Mask_model):

    vid = cv2.VideoCapture(video_path)

    num_frames = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))

    fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')
    out = cv2.VideoWriter('output.mp4', fourcc, 30.0,
                          (1280,720))
    mot_tracker = Sort() 

    frames=0
    start = time.time()
    while vid.isOpened():
        ret, frame = vid.read()
        if ret :
            img_bgr=frame
            img_rgb = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
            preds, = FaceNet_model.predict(img_rgb, det_threshold=.9, iou_threshold=.5)

            if len(preds['boxes']) !=0:

                detections=get_detections(img_rgb,preds,Mask_model)
                detections.reshape(len(preds['boxes']),-1)
                tracked_objects = mot_tracker.update(detections)
                img_bgr=draw_bounding_boxes(img_bgr,tracked_objects)

            out.write(img_bgr)
            frames += 1
            print("FPS of the video processing is {:5.2f} completed {:2.5f} %".format( frames / (time.time() - start),(frames/num_frames)*100))
        else: 
          break   
    out.release()
    vid.release()
```
